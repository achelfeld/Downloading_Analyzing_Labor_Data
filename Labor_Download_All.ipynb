```python
import requests
import urllib.request
import time
import shutil
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
from pandas import *
```

```python
#This is the URL of the website with the data
url = 'https://www.bls.gov/oes/tables.htm' 
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup.findAll('a')
```

```python
one_a_tag = soup.findAll('a')[210]
link = one_a_tag['href']
link #Just to see what we are looking at
```

```python
#The links for all data from 2004-2018 have "special.requests/oesm" in them
all_files=soup.find_all(href=re.compile("special.requests/oesm"))
len(all_files)
```

```python
#Let's just see what the 7th element contains
tag = soup.find_all(href=re.compile("special.requests/oesm"))[7]
link = tag['href']
file_name=link[23:34]
file_name #Check to see if this is the file we want to download
```

```python
#After doing some exploring using soup.find_all,
#I figured out which files we want to download
#and this code loops through the file names
for x in range(10,72):
    tag = soup.find_all(href=re.compile("special.requests/oesm"))[x]
    link = tag['href']
    url = 'https://www.bls.gov'+ link
    file_name=link[23:30]+'.zip'
    # Download the file from `url` and save it locally under `file_name`:
    with urllib.request.urlopen(url) as response, open(file_name, 'wb') as out_file: 
        shutil.copyfileobj(response, out_file)
    time.sleep(1) #pause the code for a sec
```
